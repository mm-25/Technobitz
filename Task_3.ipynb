```python
tips_df = sns.load_dataset("tips")
```

```python
sns.violinplot(x="total_bill", y="sex",
               data=tips_df)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex" , data=tips_df)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips_df)
```

```python
sns.catplot(x="time", y="total_bill", hue="smoker", data=tips_df)
```

```python
sns.countplot(x="smoker", hue="sex" , data=tips_df)
```

```python
sns.swarmplot(x=tips_df["tip"])
```

```python
sns.catplot(x="time" , y="tip" , hue="smoker" , data=tips_df)
```

```python
sns.countplot(y="day", hue="smoker", data=tips_df)
```
